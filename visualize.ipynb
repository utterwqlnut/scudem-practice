```python
import numpy as np
import matplotlib.pyplot as plt
from projectile import Projectile
from dog import Dog
```

```python
ball = Projectile(5,10,0.1,0.25,10,5)
x,y = ball.simulate_points(100,4)
```

```python
plt.plot(np.linspace(0,4,100),y,label='y')
plt.plot(np.linspace(0,4,100),x,label='x')
plt.plot([0]*5,label='0 horizontal dist')
plt.legend()
```

```python
d = Dog(ball,0,0.377,0.376)
plt.plot(np.linspace(0,4,100),d.get_positions(100,4)[1],label='True Y position')
plt.plot(np.linspace(0,4,100),y,label='Dog Predicted Y position')
plt.legend()
```

```python
d.e=0.0
thetas = d.get_angles(100,4)
plt.plot(np.linspace(0,4,100)[:30],thetas[:30],label='Ideal Dog Head Angle')
d.e = 5.0
thetas = d.get_angles(100,4)
plt.plot(np.linspace(0,4,100)[:30],thetas[:30],label='Dog Head Angle with Positive Error')
d.e = -5.0
thetas = d.get_angles(100,4)
plt.plot(np.linspace(0,4,100)[:30],thetas[:30],label='Dog Head Angle with Negative Error')
plt.legend()
```
